#Task 6. What is the most popular family name in the dataset (1 pt)

```python
df['FamilyName'] = df['Name'].str.split(',').str[0]
family_name_counts = df['FamilyName'].value_counts()

most_popular_family_name = family_name_counts.index[0]
cnt = family_name_counts.values[0]

print("The most popular family name is:", most_popular_family_name)
print("Occurrences:", cnt)
```

##Task 7. What is the mean and std for all columns for survived male passengers (1 pt)

```python
survived = df[(df['Sex'] == 'male') & (df['Survived'] == 1)]

print("The mean for all columns for survived male passengers:")
print(survived.mean(numeric_only=True))
print("\n---------------------------------\n")
print("The standard deviation for all columns for survived male passengers:")
print(survived.std(numeric_only=True))
```

##Task 10 Implement simple model to predict whether the person have survived according to knowledge you got from data analysis. Then calculate accuracy of your predictions on the dataset.
```python
df['predict_survived'] = ((df['Sex'] == 'female') |
 ((df['Sex'] == 'male') & (df['Age'] > 50) & (df["Fare"] >= 90) & (df['Embarked'] == 'S') & ((df["SibSp"] == 1) | (df["SibSp"] == 0)))).astype(int)


predictions = (df["predict_survived"] == df["Survived"]).sum()
accuracy = predictions / len(df)
print(accuracy)
```
